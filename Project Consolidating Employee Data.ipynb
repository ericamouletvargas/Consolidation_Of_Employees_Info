```python
import pandas as pd

#Storing office addresses
office_addresses = "datasets/office_addresses.csv"

#Storing employee addresses & emergency contact
employee_information = "datasets/employee_information.xlsx"

#Storing employee role
employee_role = "datasets/employee_roles.json"

#Put office_addresses into a dataframe
df_office_addresses = pd.read_csv(office_addresses, sep= ",")

#We print the head to understand better the data
print(df_office_addresses.head())

#Put employee addresses into a dataframe
df_employee_adresses = pd.read_excel(employee_information, sheet_name= "employee_addresses", header = 0 )

#We print the head to understand better the data
print(df_employee_adresses.head())

#Put employee emergency contacts into a dataframe
df_employee_emergency_contacts = pd.read_excel(employee_information, sheet_name= "emergency_contacts", header = None )

#Declare the variables for the headers
header_emergency_contact=['employee_id', 'last_name', 'first_name', 'emergency_contact', 'emergency_contact_number','relationship']

#Add the header to the table of emergency contact
df_employee_emergency_contacts.columns = header_emergency_contact

#We print the head to understand better the data
print(df_employee_emergency_contacts.head())

#Put employee role into a dataframe
df_employee_role = pd.read_json(employee_role, orient = 'index')


df1=pd.DataFrame(df_employee_role)
#We print the head to understand better the data
print(df1.head())

#We start merging the tables together by part; we first start with employee adresses and emergency contact
df_employees_general_info = df_employee_adresses.merge(df_employee_emergency_contacts, how = "left", on = "employee_id")

#We then have a look at the merged data
print(df_employees_general_info.head())

#We then have to merge this new data_frame to office adresses
df_employees_info_with_office = df_employees_general_info.merge(df_office_addresses, how = "left", right_on = "office_country", left_on = "employee_country")

#We then have a look at the merged data
print(df_employees_info_with_office.head())

#We then have to merge this new data_frame to office adresses
df_employees_full_info = df_employees_info_with_office.merge(df_employee_role, how = "left", right_on = df_employee_role.index, left_on = "employee_id")

#We then have a look at the merged data
print(df_employees_full_info.head())

#We want to know how many NaN the following table has
df_employees_full_info.isna().sum()

#We change the NaN values in office to the word remote
df_employees_with_remote = df_employees_full_info.fillna(value = "Remote")

#We have a look at the new dataframe
print(df_employees_with_remote.head())

#We check if there is still NaN in our dataframe
df_employees_with_remote.isna().sum()

#and we reordered the columns to get the dataframe we want
employees_almost_final = df_employees_with_remote.reindex(columns=["employee_id", "first_name", "last_name", "employee_country", "employee_city", "employee_street", "employee_street_number", "emergency_contact", "emergency_contact_number", "relationship", "monthly_salary", "team", "title", "office", "office_country", "office_city", "office_street", "office_street_number"])

#We have a look at the data
print(employees_almost_final.head())

# We then set the index to the data frame to employee id
employees_final = employees_almost_final.set_index('employee_id')

#We check that the index was rightly set
print(employees_final.head())
```
